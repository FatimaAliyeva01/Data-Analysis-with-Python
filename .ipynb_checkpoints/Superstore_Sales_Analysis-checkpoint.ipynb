### Customer segmentation

```python
types_of_customers = df['Segment'].unique()
print(types_of_customers)
# In our project, we are identifying and displaying unique values in the 'Segment' column of the DataFrame.

# Extracting Unique Customer Segments:
# The code 'types_of_customers = df['Segment'].unique()' isolates the unique values present in the 'Segment' column.
# The '.unique()' method returns an array containing distinct elements from the specified column.
# Purpose:
# Understanding the unique values within the 'Segment' column is essential for segmentation analysis or categorization.
# It allows us to identify the distinct customer segments present in the dataset, providing insights into the diversity of customers.
```

```python
number_of_customers = df['Segment'].value_counts().reset_index()
print(number_of_customers)
# In our project, we are determining the count of customers in each unique segment and displaying the results.

# Counting Number of Customers in Each Segment:
# The code 'number_of_customers = df['Segment'].value_counts().reset_index()' calculates the number of customers in each unique segment of the 'Segment' column.
# The '.value_counts()' method counts the occurrences of each unique value, and 'reset_index()' is used to convert the result into a DataFrame for better readability.
# Purpose:
# This operation is useful for understanding the distribution of customers across different segments.
# It provides a clear overview of how many customers belong to each unique segment, aiding in segmentation analysis and decision-making.
```

```python
number_of_customers=number_of_customers.rename(columns={'index':'Customer types','Segment':'Total Customers'})
print(number_of_customers)
# In our project, we are enhancing the clarity of the customer segmentation analysis results by renaming the columns.

# Renaming Columns for Improved Readability:
# The code 'number_of_customers = number_of_customers.rename(columns={'index':'Customer types','Segment':'Total Customers'})' is modifying the column names in the 'number_of_customers' DataFrame.
# It renames the 'index' column to 'Customer types' and the 'Segment' column to 'Total Customers' for better interpretability.
# Purpose:
# Renaming columns enhances the clarity of the presented information, making it easier for stakeholders to understand the content of each column.
# In this case, it provides a clearer representation of customer types and the total number of customers in each segment.
```

```python
plt.pie(number_of_customers['Total Customers'],labels=number_of_customers['Customer types'],autopct='%1.1f%%')
plt.title('Distribution of Customers')
plt.show()
# In our project, we are visualizing the distribution of customers across different segments using a pie chart.

# Creating a Pie Chart for Customer Distribution:
# The code 'plt.pie(number_of_customers['Total Customers'], labels=number_of_customers['Customer types'], autopct='%1.1f%%')' utilizes the 'pyplot' module from the 'matplotlib' library to generate a pie chart.
# It takes the total number of customers in each segment ('Total Customers') as values, assigns the unique customer types ('Customer types') as labels, and includes percentage annotations using the 'autopct' parameter.

# Purpose:
# This visualization provides a clear representation of the proportion of customers in each segment relative to the total customer population.
# Pie charts are effective for conveying the distribution of categorical data in a visually appealing way.
```

```python
sales_category=df.groupby('Segment')['Sales'].sum().reset_index()
sales_category=sales_category.rename(columns={'Segment':'Customer Type','Sales':'Total Sales'})
print(sales_category)
# In our project, we are conducting a sales analysis by grouping the DataFrame based on customer segments.

# Grouping by Customer Segments and Calculating Total Sales:
# The code 'sales_category = df.groupby('Segment')['Sales'].sum().reset_index()' uses the 'groupby()' method to group the DataFrame by the 'Segment' column.
# It then calculates the sum of sales for each customer segment using the '['Sales'].sum()' expression.
# Finally, 'reset_index()' is applied to convert the result into a DataFrame for better readability.

# Purpose:
# This analysis helps understand the contribution of each customer segment to the overall sales, providing insights into sales distribution across different customer types.
```

```python
plt.bar(sales_category['Customer Type'], sales_category['Total Sales'])
plt.title('Sales Customer Category')
plt.xlabel('Customer Type')
plt.ylabel('Total Sales')
plt.show()
# In our project, we are visualizing the total sales across different customer segments using a bar chart.

# Creating a Bar Chart for Sales by Customer Category:
# The code 'plt.bar(sales_category['Customer Type'], sales_category['Total Sales'])' utilizes the 'pyplot' module from the 'matplotlib' library to generate a bar chart.
# It takes the 'Customer Type' column as the x-axis labels, 'Total Sales' column as the corresponding y-axis values, and creates bars for each customer category.

# Purpose:
# This visualization provides a clear comparison of total sales across different customer segments.
# Bar charts are effective for conveying the magnitude of values and identifying trends or disparities between categories.
```

# 6. Sales Analysis

```python
# Convert the "Order Date" column to datetime format

df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst=True)

# Group the data by years and calculate the total sales amount for each year

yearly_sales = df.groupby(df['Order Date'].dt.year)['Sales'].sum()
yearly_sales = yearly_sales.reset_index()
yearly_sales = yearly_sales.rename(columns={'Order Date': 'Year', 'Sales':'Total Sales'})
print(yearly_sales)
#The code provides a summary of the total sales for each year, allowing stakeholders to analyze the sales trends over time.
```

```python
# Ploting a bar graph

plt.bar(yearly_sales['Year'], yearly_sales['Total Sales'])

# Labels
plt.title('Yearly Sales')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)

plt.show()
#The code aims to visually represent the yearly sales using a bar graph.
#The graph provides a clear view of the total sales for each year, allowing stakeholders to observe trends and variations.
```

```python
# Create a line graph for total sales by year
plt.plot(yearly_sales['Year'], yearly_sales['Total Sales'], marker='o', linestyle='-')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.title('Total Sales by Year')

# Display the plot
plt.tight_layout()

plt.show()
#The purpose of this code is to visually represent the total sales by year using a line graph.
#The graph helps stakeholders observe trends and variations in sales across different years.
```

```python
# Convert the "Order Date" column to datetime format
df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst=True)

# Filter the data for the year 2018
year_sales = df[df['Order Date'].dt.year == 2018]

# Calculate the quarterly sales for 2018
quarterly_sales = year_sales.resample('Q', on='Order Date')['Sales'].sum()

quarterly_sales = quarterly_sales.reset_index()
quarterly_sales = quarterly_sales.rename(columns={'Order Date': 'Quarter', 'Sales':'Total Sales'})

print("Quarterly Sales for 2018:")
print(quarterly_sales)
```

```python
# Create a line graph for total sales by year
plt.plot(quarterly_sales['Quarter'], quarterly_sales['Total Sales'], marker='o', linestyle='--')

plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.title('Total Sales by Year')

# Display the plot
plt.tight_layout()
plt.xticks(rotation=75)

plt.show()
```

```python
# Convert the "Order Date" column to datetime format
df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst=True)

# Filter the data for the year 2018
year_sales = df[df['Order Date'].dt.year == 2018]

# Calculate the monthly sales for 2018
monthly_sales = year_sales.resample('M', on='Order Date')['Sales'].sum()

# Renaming the columns
monthly_sales = monthly_sales.reset_index()
monthly_sales = monthly_sales.rename(columns={'Order Date':'Month', 'Sales':'Total Montly Sales'})

# Print the monthly and quarterly sales for 2018
print("Monthly Sales for 2018:")
print(monthly_sales)
```

```python
# Create a line graph for total sales by year
plt.plot(monthly_sales['Month'], monthly_sales['Total Montly Sales'], marker='o', linestyle='--')

plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.title('Total Sales by Month')

# Display the plot
plt.tight_layout()
plt.xticks(rotation=75)

plt.show()
#The purpose of this code is to visually represent the total monthly sales by year using a line graph.
#The graph helps stakeholders observe trends and variations in sales across different months.
```

# 7. Sales Trends

```python
# Assuming you have a DataFrame named "df" with columns "Order Date" and "Sales amount"

# Convert the "Order Date" column to datetime format
df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst=True)

# Group the data by quarters and calculate the total sales amount for each quarter
quarterly_sales = df.groupby(df['Order Date'].dt.to_period('Q'))['Sales'].sum()

# Plot the sales trends for months, quarters, and years
plt.figure(figsize=(12, 20))

# Quarterly Sales Trend
plt.subplot(3, 1, 2)
quarterly_sales.plot(kind='line', marker='o')
plt.title('Quarterly Sales Trend')
plt.xlabel('Quarter')
plt.ylabel('Sales Amount')

# Adjust layout and display the plots
#plt.tight_layout()
plt.show()
#The code aims to visualize the quarterly sales trend using a line graph, providing insights into sales patterns over different quarters.
```

```python
# Convert the "Order Date" column to datetime format
df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst=True)

# Group the data by months and calculate the total sales amount for each month
monthly_sales = df.groupby(df['Order Date'].dt.to_period('M'))['Sales'].sum()

# Plot the sales trends for months
plt.figure(figsize=(12, 26))

# Monthly Sales Trend
plt.subplot(3, 1, 1)
monthly_sales.plot(kind='line', marker='o')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Sales Amount')

# Adjust layout and display the plots
# plt.tight_layout()
plt.show()
```
